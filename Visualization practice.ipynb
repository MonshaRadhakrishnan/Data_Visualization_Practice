## Bar Charts and Heatmaps

```python
flight_file = '../2. Visualization Practice Datasets/flight_delays.csv'
flt = pd.read_csv(flight_file, index_col='Month', parse_dates=True)
```

```python
flt.shape
```

```python
flt
```

```python
plt.figure(figsize=(7,5))
plt.title('Avg arrival delay of Spirit airlines')
sns.barplot(x=flt.index, y=flt['NK'])
plt.ylabel('Spirit Airlines - Arrival delay')
plt.show()
```

## HeatMap

```python
plt.figure(figsize=(7,5))
plt.title('Arrival Delay for each airline, by month')
sns.heatmap(data=flt, annot=True)
plt.xlabel('Airline')
plt.show()
```

## Scatter Plots

```python
insurance_file = '../2. Visualization Practice Datasets/insurance.csv'
ins = pd.read_csv(insurance_file)
```

```python
ins.head()
```

```python
sns.scatterplot(x=ins['bmi'], y=ins['charges']) ## more bmi more charges
plt.show()
```

### Adding regression line to get the best fit line using - regplot

```python
sns.regplot(x=ins['bmi'], y=ins['charges']) ## regplot --> plots the best fit line for the data
```

```python
sns.scatterplot(x=ins['bmi'], y=ins['charges'], hue=ins['smoker'])
## can use another variable to better understand by giving it as a hue
## Color diff to differenciate the smokers vs non smokers
```

```python
## it is clear now that Smokers with high or low bmi have to pay more than Non-Smokers
```

## Now we can add regression lines for both using lmplot

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=ins)
```

```python
## we can see line for smokers is steeper than non-smokers
```

```python
sns.swarmplot(x=ins['smoker'], y=ins['charges'])
plt.show()
```

#### We can clearly infer that 

### On an avg non-smokers are charged less than smokers
### Customers who pay more are smokers and customers who pay less are non-smokers
